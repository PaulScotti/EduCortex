```python
import numpy as np
import matplotlib.pyplot as plt
import json
import requests
from tqdm import tqdm
import time
```

```python
# associations = np.full((22750,20),'NAN')
# associations = {}
for x in range(-60,61,5):
    for y in range(-100,71,5):
        print("x{} y{}".format(x,y))
        for z in range(-60,70,5):
            if x<45:
                continue
            if x==45 and y<-15:
                continue
            time.sleep(20)
            page = requests.get("https://neurosynth.org/api/locations/compare/?x={}&y={}&z={}&r=6".format(x,y,z))
            if len(page.content)==141:
                continue
            data = json.loads(page.content)['data']
            in0 = np.flip(np.sort(np.array(data)[:,1]))[:20]
            in0 = in0[in0!='0.0']
            ind = np.flip(np.argsort(np.array(data)[:,1]))[:20]
            ind = ind[:len(in0)]
            associations['x{}y{}z{}'.format(x,y,z)] = list(np.array(data)[ind,0])
```

```python
np.save('associations',associations)
```

```python
with open('assocations.json', 'w') as f:
    json.dump(associations, f)
```
